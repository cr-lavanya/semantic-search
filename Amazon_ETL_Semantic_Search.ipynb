```python
df = pd.read_csv("amazon.csv")
```

```python
df = df[['product_id', 'product_name', 'about_product']]
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
def preprocess_text(text):
  """Convert text to lower case and remove extra spaces"""
  return text.lower().strip()
```

```python
df["about_product"] = df["about_product"].apply(preprocess_text)
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
text = df["about_product"].tolist()
```

```python
embeddings = model.encode(
    text,
    convert_to_numpy=True,
    show_progress_bar=True
)
```

```python
import faiss

dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)

print("Total vectors:", index.ntotal)
```

```python
def semantic_search(query, k=5):
    query_embedding = model.encode([query], convert_to_numpy=True)
    distances, indices = index.search(query_embedding, k)
    return df.iloc[indices[0]][["product_id", "product_name", "about_product"]]
```

```python
semantic_search("wireless bluetooth headphones")
```
